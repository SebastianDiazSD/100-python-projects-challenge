## 🔹 5. Sorting Values & Adding Columns: Majors with the Most Potential vs Lowest Risk

### Lowest Risk Majors

To measure salary risk, we calculate the difference between the 90th percentile and 10th percentile salaries.

A smaller spread = lower income uncertainty.  
A larger spread = higher variability and risk.

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

### Sorting by the Lowest Spread

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

### Top 5 degrees with the highest values in the 90th percentile. 

```python
top_5 = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
top_5[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

### Majors with the Greatest Spread in Salaries

```python
highest_spread = clean_df.sort_values('Spread', ascending=False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```

Notice how 3 of the top 5 are present in both. This means that there are some very high earning Economics degree holders out there, but also some who are not earning as much

## 🔹 6. Grouping and Pivoting Data with Pandas

To better understand salary differences by academic field, we group the data by category:
- STEM
- Business
- HASS

```python
clean_df.groupby('Group').count()
```

### Average salary by group

For better readability it is used: pd.options.display.float_format = '{:,.2f}'.format

```python
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group').mean(numeric_only=True)
```

STEM majors show the highest average mid-career salaries.
HASS majors have the lowest average starting salaries.
Business majors show competitive 90th percentile earnings.

## 🧠 Final Conclusions

- STEM degrees dominate mid-career earnings.
- Business majors show strong high-end earning potential.
- HASS majors generally start with lower salaries.
- Some majors (like Economics and Finance) show very high salary spread, meaning high risk but also high reward.
